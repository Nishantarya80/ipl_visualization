# Max wickets

```python
#cleaning
wicketdata= ipl_BBB_data[(ipl_BBB_data.Dissimal_Type!=" ") 
                         &(ipl_BBB_data.Dissimal_Type!="run out")
                        &(ipl_BBB_data.Dissimal_Type!="retired hurt")] 
```

```python
#grouping sorting sclicing
wicket=wicketdata.groupby(['Bowler_Id']).count()
wicket=wicket.sort_values(by=['Dissimal_Type'], ascending=False)
wicket=wicket.loc[:,['Dissimal_Type']]
wicket=wicket.reset_index()
```

```python
#merging 
player=player_data.rename(index=str, columns={"Player_Id": "Bowler_Id"})
player=player.drop(['Is_Umpire', 'Unnamed: 7'], axis=1)
wicket.merge(player,on='Bowler_Id').rename(index=str, columns={"Dissimal_Type": "Wickets"})
```

# Max Runs

```python
Rundata=ipl_BBB_data[(ipl_BBB_data.Batsman_Scored!=" ") 
                         &(ipl_BBB_data.Batsman_Scored!="Do_nothing")
                        &(ipl_BBB_data.Batsman_Scored!="0")]
```

```python
Rundata.Batsman_Scored = Rundata.Batsman_Scored.astype(int)
Rundata=Rundata.groupby(["Striker_Id"]).sum().sort_values(by=['Batsman_Scored'], ascending=False)
Rundata=Rundata.reset_index()
```

```python
player=player_data.rename(index=str, columns={"Player_Id": "Striker_Id"})
player=player.drop(['Is_Umpire', 'Unnamed: 7'], axis=1)
Rundata.merge(player,on='Striker_Id')
```

# Max Six 

```python
Sixdata=ipl_BBB_data[(ipl_BBB_data.Batsman_Scored=="6")]
Sixdata=Sixdata.groupby(["Striker_Id"]).count().loc[:,['Batsman_Scored']].sort_values(by=['Batsman_Scored'], ascending=False)
Sixdata=Sixdata.reset_index()
```

```python
player=player_data.rename(index=str, columns={"Player_Id": "Striker_Id"})
player=player.drop(['Is_Umpire', 'Unnamed: 7'], axis=1)
Sixdata.merge(player,on='Striker_Id').rename(index=str, columns={"Batsman_Scored": "No. of Sixs"})
```

# Max Four

```python
Fourdata=ipl_BBB_data[(ipl_BBB_data.Batsman_Scored=="4")]
Fourdata=Fourdata.groupby(["Striker_Id"]).count().loc[:,['Batsman_Scored']].sort_values(by=['Batsman_Scored'], ascending=False)
Fourdata=Fourdata.reset_index()
```

```python
player=player_data.rename(index=str, columns={"Player_Id": "Striker_Id"})
player=player.drop(['Is_Umpire', 'Unnamed: 7'], axis=1)
Fourdata.merge(player,on='Striker_Id').rename(index=str, columns={"Batsman_Scored": "No. of Fours"})
```

# IPL season

```python
Season_data = pd.DataFrame()
match_data.Match_Date = pd.to_datetime(match_data.Match_Date)
for i in match_data.Season_Id.unique():
    Temp_Season_data=match_data[(match_data.Season_Id==i)].sort_values(by=['Match_Date'], ascending=False)
    Season_data=Season_data.append(Temp_Season_data.head(1),ignore_index=True)
```

```python
temp_team_data=team_data.rename(index=str, columns={"Team_Id": "Match_Winner_Id"})
Season_data=Season_data.merge(temp_team_data,on='Match_Winner_Id')
Season_data=Season_data.loc[:,['Season_Id','Team_Name','Match_Date','Win_Type','Won_By','Venue_Name',
                   'City_Name','Host_Country','Man_Of_The_Match_Id']]
Seasion_data=Seasion_data.reset_index()
```

```python
player=player_data.loc[:,['Player_Id','Player_Name']].rename(index=str, columns={"Player_Id": "Man_Of_The_Match_Id",
                                                                                "Player_Name": "Man_Of_The_Match"})
Season_data.merge(player,on='Man_Of_The_Match_Id').drop(['Man_Of_The_Match_Id'], axis=1)
```
